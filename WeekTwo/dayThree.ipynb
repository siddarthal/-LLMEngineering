```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr

load_dotenv(override=True)

client = OpenAI()


```

```python
MODEL = 'gpt-4.1-mini'
system_prompt = """
You are a helpful assistant that can answer questions and help with tasks.
"""
```

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    stream = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt}]+history+[{"role": "user", "content": message}],
        stream=True
    )
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        yield response




```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```

```python
system_prompt = "You are a helpful assistant for AXIS Bank , you are here to explain different polices and questions for bank\
    for the customers. For example if the customer asks about the different types of credit cards, you can tell types of \
        cards [Myzone , Rewards, Signature, Platinum, etc.]. and interest rates[18.5% - 25.5% per annum]\
            and different types of loans [Home loan, Car loan, Personal loan, etc.] and interest rates[10.5% - 18.5% per annum]\
                and different types of insurance [Life insurance, Health insurance, etc.] and interest rates[12.5% - 15.5% per annum]\
                    and different types of investment [Mutual funds, Stocks, Bonds, etc.] and interest rates[15.5% - 20.5% per annum]\
                        and different types of retirement plans [Pension, Annuity, etc.] and interest rates[18.5% - 25.5% per annum]\
                            and different types of wealth management [Wealth management, Financial planning, etc.] and interest rates[20.5% - 25.5% per annum]\
                                and different types of tax planning [Tax planning, Tax saving, etc.] and interest rates[25.5% - 30.5% per annum]\
                                    and different types of credit card offers [Credit card offers, etc.] and interest rates[25.5% - 30.5% per annum]\
                                        and you can google for more information if the customer asks for it.     "
```

```python
def chat(message, history):
    relevant_system_prompt = system_prompt
    if "customer care" in message.lower():
        relevant_system_prompt = "You are a customer care bot representative for AXIS Bank, Help customer by providing details to talk with customer care executive. "
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    stream = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": relevant_system_prompt}]+history+[{"role": "user", "content": message}],
        stream=True
    )
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        yield response
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```
