```python
import tiktoken

encoding =tiktoken.encoding_for_model("gpt-4o-mini")
```

```python
text ="hello i am siddartha whats ur name"
tokens= encoding.encode(text)
tokens
```

```python
for token in tokens:
    tokenId = encoding.decode([token])
    print(f"{tokenId} - {token}")
```

```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)

openai_key=os.getenv("OPENAI_API_KEY")

```

```python
from openai import OpenAI

openai =OpenAI()



```

```python
messages =[{"role":"system","content":"you are a helpful assistant"},
           {"role":"user","content":"MY SCHOOL IS MGIT(MAHATMA GANDHI INSTITUTE OF TECHNOLOGY) HYDERABAD"}
]

```

```python
def sendMessage(messages):
    response = openai.chat.completions.create(model="gpt-4o-mini",messages=messages)
    return response.choices[0].message.content
```

```python
sendMessage(messages)
```

```python
messages2 =[{"role":"system","content":"you are a helpful assistant"},
           {"role":"user","content":"what is my school and where is my school located"}
]

sendMessage(messages2)
```
