```python
user_prompt="for the give contet please provide a summary of content and return in document markdown format"
system_prompt="You are a helpful assitant who can summarize the content of website , you can skip all content that is breaking lines or escape charactersand return the summary in markdown format but not code markdown"
```

```python
messages=[{"role":"system",message:system_prompt},{"role":"user",message:user_prompt}]
messages
```

```python
def addMessages(websiteData):
    final_message=[{"role":"system","content":"system_prompt"},{"role":"user","content":user_prompt+websiteData}]
    return final_message
```

```python
def getSummary(url):
    dataScrapped=fetch_website_contents(url)
    response=openai.chat.completions.create(model="gpt-5-nano",messages=addMessages(dataScrapped))
    display(Markdown(response.choices[0].message.content))
```

```python
getSummary("https://edwarddonner.com")
```
